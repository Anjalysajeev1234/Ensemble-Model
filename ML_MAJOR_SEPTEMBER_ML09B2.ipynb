#### Box plots for showing trends and Outlier Identification if any :-

```python
def find_out(df,var,dist):
    IQR = df[var].quantile(0.75)-df[var].quantile(0.25)
    upper_limit = df[var].quantile(0.75)+(IQR * dist)
    lower_limit = df[var].quantile(0.25)+(IQR * dist)
    return upper_limit,lower_limit
```

### Box plot for fav_number

```python
fig, axs = plt.subplots(ncols = 2, figsize=(15,8))
sns.boxplot(df.fav_number, ax=axs[0], orient = 'v')
sns.boxplot(df.fav_number, ax=axs[1], orient = 'v', showfliers=False)
ul,ll = find_out(df,'fav_number',1.5)
x = df[df.fav_number>ul]
print("Number of Outliers =",x.shape[0])
print("Outlier Percentage =",round(x.shape[0]*100/df.shape[0],2),"%")
```

###### Thus,Outliers spotted.



### Box plot for tweet_count

# Questions and answers based on the dataset:

```python
#Q1)What are the most common words used by Males and Females?
from collections import Counter
#function to split sentences into words
def text_find(df2):
    y=[]
    for i in range(0, df2.shape[0]):
        rand = df2['text'].values[i]
        x = rand.split(' ')
        for j in x:
            y.append(j)
    c = Counter(y)
    return c
#reading the datafile
df = pd.read_csv('Information.csv')
df = df[df['gender'].notnull()]
#to lock the rows that contain gender as male
df1 = df[df['gender'].str.contains('male')]
text_list1=text_find(df1)
#to lock the rows that contain gender as female
df2 = df[df['gender'].str.contains('female')]
text_list2=text_find(df2)
for key in text_list2: 
    if key in text_list1: 
        text_list2[key] = text_list2[key] + text_list1[key] 
    else: 
        pass
print("The most common words used by Males and Females: ") 
for i in text_list2.most_common(3):
    print(i[0])
```

```python
#cleaning the text
import re
text_list = []
for text in df['text']:
    text = text.lower()
    #removing punctuation
    text = re.sub('\s\W',' ',text)  
    text = re.sub('\W\s',' ',text)
    #If there are any double spaces 
    text = re.sub('\s+',' ',text)
    # keeps only alphabets and removes other characters
    text = re.sub("[^a-zA-Z]", " ", text)
    text = "".join(text)
    text_list.append(text)
#a dictonary with 52000 words
words = open("words.txt").readlines()
words = [word.strip() for word in words]
x = []
for i in text_list:
    counter = 0
    row = i.split()
    for j in row:
        if row not in words:
            counter = counter+1
    x.append(counter)
```

```python
from pandas import DataFrame
df_typos = DataFrame (x,columns=['no_of_typos'])
gender_typos = df['gender']
gender_typos = DataFrame(gender_typos,columns=['gender'] )
df_errors = df_typos.join(gender_typos)
```

```python
#removing NaN values
df_errors = df_errors[df_errors['gender'].notna()]
df_errors
```
